Berikut adalah fungsi - fungsi dari regresi linear dengan cara membuat terlebih dahulu matrixnya, kemudian mencari koefisien dari regresi linear (pada kasus ini menyatakan a0 dan a1) menggunakan eliminasi gauss dan backsub.
Hasil yang didapatkan(secara berurutan): Matriks konstanta dan matrix koefisien, hasil solusi SPL.

```python

#fungsi ini akan mengembalikan matrix koefisien dan matrix konstanta
def my_gls(x,y,n):
    """
    x : merupakan vektor x
    y : merupakan vektor y (panjang x dan y sama)
    n : merupakan derajat dari polinom yang akan dibuat
    
    return value: sebuah array yang merupakan matrix koefsien dan matrix konstanta
    """
    lenX=len(x)
    lenY=len(y)
    
    matriks = np.zeros((n+1,n+1),dtype=float)
    koefMatriks = np.zeros((n+1,1),dtype=float)
    for i in range (0,n+1):
        for j in range (0,n+1):
            if i==0 and j==0:
                matriks[i][j] = n
            elif i==0 and j>0:
                matriks[i][j]= np.sum(x**j)
            else:
                matriks[i][j] = np.sum(x**(i+j))

    for i in range (0,n+1):
        for j in range (0,1):
            if(i==j):
                koefMatriks[i][j] = np.sum(y)
            else:
                koefMatriks[i][j] = np.sum((x**i)*y)
    return matriks, koefMatriks
    
#fungsi eliminasi gauss               
def EG(A, b):
    """
    A: matriks koefisien berukuran mxm
    b: vektor konstanta
    
    return:
    A: matriks A yang sudah menjadi segitiga atas
    b: vektor konstanta yang berubah sesuai dengan operasi yang dilakukan terhadap matriks A
    Jika A tidak dapat dibuat menjadi segitiga atas, kembalikan A dan b yang berupa matriks 0.
    Hal ini untuk penanda bahwa tidak dapat ditemukan solusi.
    """
    
    
    N = A.shape[0] #N adalah ukuran matriks A dan vektor B
    #Lengkapi fungsi ini
  
    for i in range(N-1):
        for j in range(i+1,N):
            temp = A[j][i]/A[i][i]
            A[j,:] = A[j,:] - temp * A[i,:]
            b[j] = b[j]-temp*b[i]
    return A,b

#fungsi backsub
def BackSub(A,b):
    """
    A: matriks segitiga atas
    b: vektor konstanta
    
    return:
    x: sebuah vektor yang merupakan solusi dari SPL
    """
    N = len(b)
    x = np.zeros(N)
   
    for i in range (N-1,-1,-1):
        temp = b[i]
        for j in range(i+1,N):
            temp = temp-A[i][j] * x[j]
        x[i] = temp/A[i][i]
    return x    

#fungsi solve    
def solve(A,B):
    
    """
    A: matriks koefisien berukuran mxm
    B: vektor konstanta
    
    return:
    x: sebuah vektor yang merupakan solusi dari SPL dalam hal ini a0, a1
    """
        
    res = EG(A,B)
    A1 = res[0]
    B1 = res[1]
    return (BackSub(A1,B1)) 

#fungsi predict
def my_predict(x,A):
    """
    x : merupakan vektor x
    A : vektor koefisien untuk model regresi
    
    return value: sebuah array yang merupakan hasil regresi dari vektor x
    """
    
    n = len(x)
    m = len(A)
    y = np.zeros(n)
    for i in range(n):
        for j in range(m):
            y[i] += A[j]*x[i]**j
         
    return y

#fungsi untuk mencari sum of square error
def my_Sr(ymodel, ytrue):
    """
    ymodel : vektor y hasil regresi
    ytrue  : vektor y yang sesungguhnya
    
    return value: sum of square error
    """
    n = len(ymodel)
    sr = 0
    for i in range (n):
        sr += (ymodel[i] - ytrue[i])**2
    return sr
```

Berikut ini kita akan menerapkan metode pengecekan K-fold cross validation untuk menentukan model terbaik dari ketiga orde(orde 2,3 dan 4).
Kita akan melakukan percobaan SR sebanyak K kali(dalam kasus ini kami menggunakan k = 10 karena pembagian dengan nilai 10 umum digunakan) untuk setiap masing-masing orde.
Hasil y ang kita dapatkan dari proses ini adalah Rata-rata SR terkecil yang mengindikasikan bahwa model orde tersebut adalah model yang terbaik.

```python
#fungsi ini akan memberikan orde terbaik (rata - rata orde terkecil)

# #data
x = np.array(carat)
y = np.array(price)

#inisiasi orde akhir
orde = 5

#inisiasi banyaknya pembagian
k=10

#inisiasi array untuk melakukan 3 kali percobaan
xTrain = np.zeros((orde+1,k),object)
xTes = np.zeros((orde+1,k),object)
yTrain = np.zeros((orde+1,k),object)
yTes = np.zeros((orde+1,k),object)
gls = np.zeros((orde+1,k),object)
predict = np.zeros((orde+1,k),object)
Sr = np.zeros((orde+1,k),object)
sum = np.zeros(orde+1)
average = np.zeros(orde+1)
result = np.zeros(orde-1)
order = np.zeros(orde-1,dtype = int)
counter = 0
for i in range(2,orde+1):
    for j in range(k):
        #melakukan split data untuk x,y tes dan train
        from sklearn.model_selection import train_test_split
        xTrain[i][j], xTes[i][j], yTrain[i][j],yTes[i][j] = train_test_split(x,y,train_size=110,random_state=j)

        gls[i][j] = my_gls(xTrain[i][j],yTrain[i][j],i)
        #print(gls[i][j])
        predict[i][j] = my_predict(xTes[i][j],solve(gls[i][j][0],gls[i][j][1]))
        Sr[i][j] = my_Sr(predict[i][j],yTes[i][j])
        sum[orde] += Sr[i][j]
    average[orde] = sum[orde] / k
    result[counter] = average[orde]
    order[counter] = i
    counter = counter+1
    
#cari sr mimimum
minimum = np.min(result)
min_idx = [idx for idx, val in enumerate(result)if val == minimum]
bestorder = int(order[min_idx])

print ("Orde dengan rata - rata terkecil(minimum) adalah:",order[min_idx], "dengan nilai rata - rata:",minimum)
```
